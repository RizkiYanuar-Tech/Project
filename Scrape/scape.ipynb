```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Create Path driver dan brave
browser_path = "C:/Program Files/BraveSoftware/Brave-Browser/Application/brave.exe"
driver_path = "D:/TUGAS/Project/chromedriver-win64/chromedriver.exe"

option = Options()
option.binary_location = browser_path

# Setup service
service_driver = Service(driver_path)
```

```python
# Setup
driver = webdriver.Chrome(service=service_driver, options=option)
driver.get("https://www.hoopshype.com/rankings/players/")
title = driver.title
print(f"Membukan halaman web {title}")
```

```python
# Waiting strategies
wait = WebDriverWait(driver, timeout=20)
wait.until(EC.all_of(EC.visibility_of_element_located((By.TAG_NAME, "thead")),
                                                EC.visibility_of_element_located((By.TAG_NAME, 'tbody')),
                                                EC.visibility_of_element_located((By.XPATH, "//td/a[contains(@href, 'salaries/players/')]"))
                                                ))
```

```python
data = []
link = set()
halaman = 1
while True:
    # Check apakah jumlah data > 50
    if len(data) >= 500:
        print("Sudah ada 500 data!")
        break
    player_stats = driver.find_elements(By.XPATH, "//tbody/tr") # Cari tag
    for col in player_stats: # Looping untuk semua data pemain
        if len(data) >= 500:
            print("Sudah ada 500 data!")
            break
        colom = col.find_elements(By.TAG_NAME, "td") # Cari tag
        if len(colom) >= 7: # Apakah kolom > 7?
                try:
                    # Cari data kolom awal
                    players = colom[1].find_element(By.TAG_NAME, "a")
                    
                    nama = players.text
                    url = players.get_attribute("href")
                    
                    RAT = colom[2].text
                    GP = colom[3].text
                    PPG = colom[4].text
                    RPG = colom[5].text
                    APG = colom[6].text

                    if url in link:
                        continue

                    link.add(url)

                    data.append({
                        "nama": nama,
                        "rating": RAT,
                        "gp": GP,
                        "point_per_game": PPG,
                        "rebound_per_game": RPG,
                        "assist_per_game": APG
                    })

                    print(f"Pemain {nama}, Point Per Game: {PPG}, Rebound Per Game: {RPG}, Assist Per Game: {APG}")
                
                except Exception as e:
                    print(f"Error: {e}")
    
    if len(data) >=500:
        print("Sudah ada 500 data!, pengambilan data berhenti")
        break
    
    # Tambah fungsi click button
    try:
        # Tunggu tombol next muncul
        btn_next = WebDriverWait(driver, timeout=10).until(EC.presence_of_element_located((By.XPATH, "(//button[contains(@class, 'hd3Vfp')])[last()]")))

        # Apakah tombol memiliki attribut disable
        if btn_next.get_attribute("disabled"):
            print("Sudah diujung halaman website!!")
            break
        else:
            # Scroll kebawah
            driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", btn_next)
            time.sleep(2)

            # Klik button
            driver.execute_script("arguments[0].click();", btn_next)
            halaman += 1
            print(f"Pindah halaman {halaman}")

            time.sleep(10)
    except Exception as e:
        print(f"Error: {e}")
```

```python
data_gaji = []
data_player = set()

while True:
    # Check dulu jumlah data yang sudah diambil
    if len(data_gaji) >= 500:
        print("Data sudah 500! Scrape berhenti")
        break
    
    # Jika belum 500 data, maka cari element tag tbody dan tr
    player_salaries = driver.find_elements(By.XPATH, '//tbody/tr')

    # Ambil data setiap baris
    for col in player_salaries:
        if len(data_gaji) >= 500:
            print("Data sudah 500!")
            break
        
        # Cari tag td
        coloms = col.find_elements(By.TAG_NAME, 'td')
        if len(coloms) >= 3:
                # Ambil data didalam kolom disetiap baris
                try:
                    player = coloms[1].find_element(By.TAG_NAME, 'a')
                    
                    nama = player.text
                    salaries = coloms[2].text
                    url = player.get_attribute('href')

                    if url in data_player:
                         continue
                    
                    data_player.add(url)

                    data_gaji.append({
                         'nama':nama,
                         'gaji': salaries
                    })

                    print(f"Data {nama} | {salaries}")
                    
                except Exception as e:
                     print(f"Error {e}")
    
    if len(data_gaji) >= 500:
         print("Data sudah 500, pengambilan berhenti")
         break
    
    # Pindah halaman
    try:
         # Tunggu sampai button bisa di click
         btn = wait.until(EC.element_to_be_clickable((By.XPATH, "(//button[contains(@class, 'hd3Vfp')])[last()]")))

         # Check status button
         if btn.get_attribute('disabled'):
            print("Halaman Terakhir")
            break
         else:
            # Scroll kebawah
            driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", btn)
            time.sleep(5)

            # klik tombol next
            btn.click()
            time.sleep(10)
            
            halaman += 1
            print(f"Halaman ke {halaman}")

    except Exception as e:
         print(f"Error {e}")
```
